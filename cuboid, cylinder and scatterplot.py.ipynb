```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import mpl_toolkits.mplot3d.art3d as art3d
from matplotlib.patches import Circle
import math
import timeit

start = timeit.default_timer()

def plot_3D_cylinder(radius, height, elevation=0, resolution=100, color='c', x_center = 0, y_center = 0):
    fig=plt.figure()
    ax = Axes3D(fig, azim=30, elev=30)

    x = np.linspace(x_center-radius, x_center+radius, resolution)
    z = np.linspace(elevation, elevation+height, resolution)
    X, Z = np.meshgrid(x, z)

    Y = np.sqrt(radius**2 - (X - x_center)**2) + y_center # Pythagorean theorem

    ax.plot_surface(X, Y, Z, linewidth=0, color=color)
    ax.plot_surface(X, (2*y_center-Y), Z, linewidth=0, color=color)

    floor = Circle((x_center, y_center), radius, color=color)
    ax.add_patch(floor)
    art3d.pathpatch_2d_to_3d(floor, z=elevation, zdir="z")

    ceiling = Circle((x_center, y_center), radius, color=color)
    ax.add_patch(ceiling)
    art3d.pathpatch_2d_to_3d(ceiling, z=elevation+height, zdir="z")

    ax.set_xlabel('x-axis')
    ax.set_ylabel('y-axis')
    ax.set_zlabel('z-axis')
    


   
    x= 0
    y= 0
    z= 0

    dx= 250
    dy= int(input ("Horizontal distance = ")) +80
    dz= int(input ("Height = "))
    r1= math.floor(dx/3)
    r2= math.floor(dy/3)
    r3= math.floor(dz/3)


    fig = plt.figure()
    x1=[]
    y1=[]
    z1=[]
    #ax = fig.add_subplot(111, projection='3d')
    for i in range(0,r1):
        x1.append(3*i)
    for j in range(0,r2):
        y1.append(3*j)
    for k in range(0,r3):
        z1.append(3*k)

    for a in range(0,j):
        for b in range(0,k):
            ax.scatter(x1,y1[a],z1[b],c='lawngreen',marker='o', alpha=0.01)
    ax.bar3d(x, y, z, dx, dy, dz, alpha=.3, color = 'dodgerblue')

    plt.show()

# params
radius = 50
height = 25
elevation = 0
resolution = 10
color = 'crimson'
x_center = 125
y_center = 80


plot_3D_cylinder(radius, height, elevation=elevation, resolution=resolution, color=color, x_center=x_center, y_center=y_center)

stop = timeit.default_timer()

print('Time: ', stop - start) 
```
